# Teste Requisição async

```python
import asyncio
import aiohttp
```

```python
def payload1(numeroProcesso):

  buildPayload = json.dumps({
      "size": 100,
      "query":
      {
          "bool":
          {
              "must":
              [
                  {"match": {"numeroProcesso": f"{numeroProcesso}"}}
              ]
          }
      },
      "sort":
      [
          {
              "dataHoraUltimaAtualizacao":
              {
                  "order": "desc"
              }
          }
      ]
  })

  return buildPayload
```

```python
async def obterStatusProcesso(session, numeroProcesso):
  async with session.get(url, headers=headers, data=payload1(numeroProcesso)) as response:
    if response.status == 200:
      data = await response.json()
      print(numeroProcesso)
      return data
    else:
      print(f"Erro ao obter status do processo {numeroProcesso}. Código de status: {response.status}")
      return None

async def main():
  listaProcessos = ["10000012520255020029", "10013836720245020262", "10019474520245020715"]
  async with aiohttp.ClientSession() as session:
    tasks = [obterStatusProcesso(session, numeroProcesso) for numeroProcesso in listaProcessos]
    results = await asyncio.gather(*tasks)
    return results

result = await main()
print(result)
```

# Teste em procurar os IDs em todas as Rotas Trabalhistas

```python
url = "https://api-publica.datajud.cnj.jus.br/api_publica_trt2/_search"
```

```python
async def obterStatusProcesso(session, numeroProcesso, rota):
  async with session.get(f"https://api-publica.datajud.cnj.jus.br/api_publica_trt{rota}/_search", headers=headers, data=payload1(numeroProcesso)) as response:
    if response.status == 200:
      data = await response.json()
      print(numeroProcesso)
      return {"Processo": numeroProcesso, "Rota": rota, "Data": data}
    else:
      print(f"Erro ao obter status do processo {numeroProcesso}. Código de status: {response.status}")
      return None

async def main():
  listaProcessos = ["10000012520255020029", "10013836720245020262", "10019474520245020715"]

  listaRotas = [1, 2, 3, 4]

  async with aiohttp.ClientSession() as session:
    tasks = []

    for numeroProcesso in listaProcessos:
      for rota in listaRotas:
        print(rota)
        tasks.append(obterStatusProcesso(session, numeroProcesso, rota))

    results = await asyncio.gather(*tasks)
    return results

result = await main()

with open("resultado.json", "w", encoding="utf-8") as f:
  json.dump(result, f, indent=4, ensure_ascii=False)



print(result)
```
